```python
flights = sns.load_dataset("flights")
flights = flights.pivot("month", "year", "passengers")
ax = sns.heatmap(flights, cmap="BrBG_r")
```

```python
flights = sns.load_dataset("flights")
flights = flights.pivot("month", "year", "passengers")



for colormap in ["rocket", "BrBG_r", "BuGn", "rocket_r", "hot_r", "magma_r", "plasma_r"]:
    fig = plt.figure()
    ax = sns.heatmap(flights, cmap=colormap)
```
